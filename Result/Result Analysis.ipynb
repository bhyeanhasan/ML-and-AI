```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data  = pd.read_csv('all_result.csv')
data
```

```python
top = data.sort_values(by='CGPA', ascending=False)
topper = sns.barplot(top.head(10), x='Student Name', y='CGPA',color='orange')
topper.set_xticklabels(topper.get_xticklabels(), rotation=50, ha="right")
plt.title("Batch Topper according to CGPA (8th Final)")
plt.show()
```

```python
top_6th = data.sort_values(by='eighth', ascending=False)
sixth = sns.barplot(top_6th.head(10), x='Student Name', y='eighth',color='orange')
sixth.set_xticklabels(sixth.get_xticklabels(), rotation=40, ha="right")
plt.title("8th semester Topper")
plt.show()
```

```python
data['inc'] = data.eighth - data.seventh
dif = data.sort_values(by='inc', ascending=False)
jump = sns.barplot(dif.head(10), x='Student Name', y='inc')
jump.set_xticklabels(jump.get_xticklabels(), rotation=40, ha="right")
plt.title("Top Changes  from previous Semester")
plt.ylabel("Change in CGPA")
plt.show()
```

```python
data['max_jump'] = None


roll = np.array(data.Roll)
for id in roll:
    i = data[data['Roll'] == id]
    ok = np.array([i['first'], i['second'], i['third'], i['fourth'], i['fifth'], i['sixth'], i['seventh'], i['eighth']])
    ok = ok.reshape(-1, 8)
    i.max_jump = (max(ok[0]) - min(ok[0]))
    data[data['Roll'] == id] = i

# outcome = outcome.sort_values(by='max_jump',ascending=False)
data
```

```python
dif = data.sort_values(by='max_jump', ascending=False)
jump = sns.barplot(dif.head(10), x='Student Name', y='max_jump')
jump.set_xticklabels(jump.get_xticklabels(), rotation=40, ha="right")
plt.title("Top Jump among All Semester")
plt.ylabel("Change in CGPA")
plt.show()
```

```python
data = data.sort_values(by='max_jump', ascending=True)
lg = sns.barplot(data.head(11), x='Student Name', y='max_jump', color='orange')
lg.set_xticklabels(lg.get_xticklabels(), rotation=40, ha="right")
# plt.title("Longest Jump Ever (2nd to 6th)")
plt.title("Top Constant Result")
# plt.ylabel("Highest GPA - Lowest GPA")
plt.ylabel("GPA Variation")
plt.show()
```
