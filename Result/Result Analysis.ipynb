```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data  = pd.read_csv('all_result.csv')
data
```

```python
top = data.sort_values(by='CGPA', ascending=False)
topper = sns.barplot(top.head(16), x='Student Name', y='CGPA',color='orange')
topper.set_xticklabels(topper.get_xticklabels(), rotation=40, ha="right")
plt.title("Batch Topper according to CGPA")
plt.show()
```

```python
top_6th = data.sort_values(by='seventh', ascending=False)
sixth = sns.barplot(top_6th.head(13), x='Student Name', y='seventh',color='orange')
sixth.set_xticklabels(sixth.get_xticklabels(), rotation=40, ha="right")
plt.title("7th semester Topper")
plt.show()
```

```python
data['inc'] = data.seventh - data.sixth
dif = data.sort_values(by='inc', ascending=True)
jump = sns.barplot(dif.head(10), x='Student Name', y='inc')
jump.set_xticklabels(jump.get_xticklabels(), rotation=40, ha="right")
plt.title("Top Changes  in 7th Semester")
plt.ylabel("Change in CGPA")
plt.show()
```

```python
data['max_jump'] = None


roll = np.array(data.Roll)
for id in roll:
    i = data[data['Roll'] == id]
    ok = np.array([i['first'], i['second'], i['third'], i['fourth'], i['fifth'], i['sixth'], i['seventh']])
    ok = ok.reshape(-1, 7)
    i.max_jump = (max(ok[0]) - min(ok[0]))
    data[data['Roll'] == id] = i

# outcome = outcome.sort_values(by='max_jump',ascending=False)
data
```
