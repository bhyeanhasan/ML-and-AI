```python
course_by_gpa = df.groupby('Grade').size().reset_index(name='Count')
course_by_gpa
```

```python
# Choose a new colormap (e.g., 'Set3')
colors = plt.cm.Set3.colors[:len(course_by_gpa)]

plt.figure(figsize=(8, 8))
plt.pie(
    course_by_gpa['Count'],
    labels=course_by_gpa['Grade'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)

plt.title('Distribution of Courses by Grade Point', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Custom function to display both count and percentage
def autopct_format(pct, allvals):
    count = int(round(pct / 100. * sum(allvals)))
    return f"{pct:.1f}%\n({count})"

# Prepare data
counts = a_plus_by_semester['A+ Count']
labels = a_plus_by_semester['Semester']
colors = plt.cm.Set3.colors[:len(counts)]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    counts,
    labels=labels,
    autopct=lambda pct: autopct_format(pct, counts),
    startangle=140,
    colors=colors
)

plt.title('Distribution of A+ Grades by Semester', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

semester_stats = df.groupby('Semester').agg({
    'Weighted Point': 'sum',
    'Credit Hours': 'sum'
}).reset_index()

semester_stats['CGPA'] = semester_stats['Weighted Point'] / semester_stats['Credit Hours']

# Sort semesters if needed

# Plot
plt.figure(figsize=(10, 4))
plt.plot(semester_stats['Semester'], semester_stats['CGPA'], marker='o', linestyle='-', color='red')
plt.ylim(3.0, 3.5)  # Start Y-axis from 3.0
plt.title("Semester-wise CGPA")
plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.grid(True)
plt.tight_layout()
plt.show()
```
