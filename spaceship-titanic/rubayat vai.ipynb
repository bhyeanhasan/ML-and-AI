```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('raw.csv')
tem = data['Temperature\n (deg C)']
date = data['Date']
date = date.dropna()
date = date.to_numpy()
date
```

```python
arr = tem.to_numpy()
arr
ok = np.split(arr,61)
day=0
ds = []
for ar in ok:
    print("Day ",day,' ','Max: ',max(ar)," ",'Min: ',min(ar)," ",'AVG: ',(max(ar)+min(ar))/2)
    lis = [date[day],max(ar),min(ar),(max(ar)+min(ar))/2]
    ds.append(lis)
    day+=1
    print(lis)
print(ds)
```

```python
np.savetxt("rubaitvai.csv",
           ds,
           delimiter =", ",
           fmt ='% s')
# df = pd.DataFrame(out)
# df.to_csv('output.csv',index=False)
```
