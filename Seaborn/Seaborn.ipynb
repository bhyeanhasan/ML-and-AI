```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('penguins')
data = data.head(100)
data.head()
```

Line Plot

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data)
plt.show()
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data,hue = 'sex',palette='hls',markers=["o",">"])
plt.title('Line Plot')
plt.show()
```

Histogram

```python
sns.displot(data.bill_depth_mm,bins = [16,17,18,19,20,21,22],kde = True)
plt.show()
```

```python
sns.barplot(data= data)
```
