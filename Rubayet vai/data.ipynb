```python
from pyeto import fao
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('rubaitvai.csv')
data
```

```python
sol_dec = []
day = data.day_of_the_year
day = np.array(day)

for i in day:
    sd = fao.sol_dec(i)
    sol_dec.append(sd)

df = pd.DataFrame({'sol_dec': sol_dec})
data.sol_dec = df
data
```

```python
avp = []
tmin = data.Min
tmin = np.array(tmin)

for i in tmin:
    x = fao.avp_from_tmin(i)
    avp.append(x)
avp_df = pd.DataFrame({'avp': avp})

data.avp = avp_df
data
```

```python
sunset_hour_angle = []
ladi = 1.288582

for i in sol_dec:
    sha = fao.sunset_hour_angle(ladi, i)
    sunset_hour_angle.append(sha)

sha_df = pd.DataFrame({'sunset_hour_angle': sunset_hour_angle})
data.sunset_hour_angle = sha_df
data
```

```python
daylight_hours = []

for i in sunset_hour_angle:
    dh = fao.daylight_hours(i)
    daylight_hours.append(dh)
dh_df = pd.DataFrame({'daylight_hours': daylight_hours})
data.daylight_hours = dh_df
data
```

```python
ird = []

for i in day:
    ird_val = fao.inv_rel_dist_earth_sun(i)
    ird.append(ird_val)

ird_df = pd.DataFrame({'ird': ird})
data.ird = ird_df
data
```

```python
et_rad = []

for i in range(61):
    etrad = fao.et_rad(ladi, sol_dec[i], sunset_hour_angle[i], ird[i])
    et_rad.append(etrad)

et_rad_df = pd.DataFrame({'et_rad': et_rad})

data.et_rad = et_rad_df
data
```

```python
sol_rad = []
sunshine_hours = data['sunshine hours']
sunshine_hours = np.array(sunshine_hours)

for i in range(61):
    solrad = fao.sol_rad_from_sun_hours(daylight_hours[i],sunshine_hours[i],et_rad[i])
    sol_rad.append(solrad)

sol_rad_df = pd.DataFrame({'sol_rad' : sol_rad})
data.sol_rad = sol_rad_df
data
```

```python
cs_rad = []
alti = 7.0
for i in range(61):
    csrad = fao.cs_rad(alti,et_rad[i])
    cs_rad.append(csrad)

cs_rad_df = pd.DataFrame({'cs_rad' : cs_rad})

data.cs_rad = cs_rad_df
data
```

```python
data.to_csv('rubaitvaiupdate.csv',index=False)
```
